```python
def calculateDensityOfPlayingPerMeasure(data, title):
    
    fig, axes = plt.subplots(nrows=1, ncols=2)
    fig.set_figheight(10)
    fig.set_figwidth(20)
    fig.suptitle('Count of notes played in each measure', fontsize=14)
   
    
    secondYValue = data['Current measure'].max()
    nu = (data['Current measure'] * data['Midi number']).mean() - data['Current measure'].mean() * data['Midi number'].mean()
    de = (data['Current measure'] ** 2).mean() - data['Current measure'].mean() * data["Current measure"].mean()
    slope = nu/de
    intercept = data['Midi number'].mean() - (slope) * data['Current measure'].mean()
    otherY = slope * secondYValue + intercept
    
    data.plot(ax=axes[0], kind="scatter", x = "Current measure", y = "Midi number", title=title)
    axes[0].plot([0,secondYValue], [intercept,otherY])
    axes[0].set_xlabel("Measure number")
    axes[0].set_ylabel("Count of notes played")
    axes[1].set_xlabel("Count of notes played")   
    axes[1].set_ylabel("Frequency of count value")   
    
    data['Midi number'].plot(ax=axes[1], kind="density", title="Frequency")
```

```python
# The above, Someday presents, that over time, things get busier. More notes are played as a solo progresses, more than
# 12 notes at a time. We can look at the same data again
# and see that over time, the average number of notes increases
# also included in this visualisation is a frequency which says....
calculateDensityOfPlayingPerMeasure(noRestData[(noRestData['Title'] == 'Someday My Prince Will Come')], 'All The Things You Are (Keith Jarrett)')
```

```python
# However, overall, it is not clear not the case, in general, number of notes the same. 
calculateDensityOfPlayingPerMeasure(noRestData, "All Data")
```

```python
# This shows us about improvisation, doesn't come up in the literature, a way to measure business, tempo
# meausre musicians of tend to play lots of notes, practical questions, 

# EXAMINE AVERAGE AMOUNT OF NOTES APPEARING ON ANY CHORD

noRestData_chord = different_melody_notes_kj
```

```python
def num_mel_notes_to_chord_type(data_frame, title, performer):
    x = data_frame[data_frame["Title"] == title].reset_index(drop=True)
    x = x[x["Performer name"] == performer]
    x = x.groupby(['Current chord type', 'Current measure']).count()
    x = x.reset_index()   
    
    note_count = x.groupby(['Current chord type']).sum()["Midi number"].to_frame()
    chord_count = x.groupby(['Current chord type']).count()["Current measure"].to_frame()
    chord_count['Total number of notes played over all chords of this type'] = note_count['Midi number']
    chord_count['Number of times chord type appears in solo'] = chord_count['Current measure']
    chord_count['Average number of notes played each time chord type is used'] = chord_count['Total number of notes played over all chords of this type'] / chord_count['Current measure']
    del chord_count["Current measure"]
    
    chord_count = chord_count[['Number of times chord type appears in solo', 'Total number of notes played over all chords of this type', 'Average number of notes played each time chord type is used' ]]
    
    return chord_count
    
```

```python
# Looking at partitions of time from beats and seconds, to look at chords, ask whether playing on certain chords
# morethan others. Do dominant chords, practical appplications. 
# IN KJ however, different chords similiar, suggests, that chord is not a driver of how notes are played in 
# Jarrett. hard to compare this as two chords per bar, but in each song shows 
chord_note_count = num_mel_notes_to_chord_type(noRestData_chord, "All The Things You Are", "Keith Jarrett")
chord_note_count['Title'] = 'All The Things You Are'
chord_note_count[['Title', 'Average number of notes played each time chord type is used']]
```
