```python
def fix_title(data_row):
    return data_row["Title"][0]

def fix_performer_name(data_row):
    return data_row["Performer name"][0]

def fix_performer_collection(data_row):
    return data_row["Performer collection"][0]

def get_opening_measure(data_row):
    return data_row["Measures"][0]

def get_closing_measure(data_row):
    return data_row["Measures"][len(data_row['Measures']) - 1]

def get_starting_position_in_measure(data_row):
    return (data_row['Locations'][0] % 1920) / 480

def get_ending_position_in_measure(data_row):
    return (data_row['Locations'][len(data_row['Locations']) - 1] % 1920) / 480

def get_pitch_range(data_row):
    return max(data_row["Midi numbers"]) - min(data_row["Midi numbers"])

def get_phrase_length(data_row):
    return len(data_row["Midi numbers"])

def get_highest_pitch(data_row):
    return max(data_row["Midi numbers"])

def get_lowest_pitch(data_row):
    return  min(data_row["Midi numbers"])

def get_number_of_different_pitches_used(data_row):
    return len(set(data_row["Midi numbers"]))

def get_different_pitches_used(data_row):
    return list(set(data_row["Midi numbers"]))

def get_longest_duration(data_row):
    return max(data_row["Durations"])

def get_shortest_duration(data_row):
    return min(data_row["Durations"])

def get_number_different_durations_used(data_row):
    return len(set(data_row["Durations"]))

def get_different_durations_used(data_row):
    return list(set(data_row["Durations"]))

def transpose_midi_numbers_to_start_on_c(data_row):
    transposed_array = []
    first_note_in_melody = data_row["Midi numbers"][0]
    for i in range(0, len(data_row["Midi numbers"])):
        transposed_array.append(data_row["Midi numbers"][i] - first_note_in_melody + 60)
        
    return transposed_array

def create_notes_without_rhythm_for_render(data_row):
  
    s = stream.Stream()
    for i in range(0, len(data_row["Midi numbers"])):
        a = note.Note(data_row["Midi numbers"][i])
        s.append(a)
    return s

def create_notes_with_rhythm_for_render(data_row):
    durations = data_row["Durations"]
    s = stream.Stream()
    for i in range(0, len(data_row["Midi numbers"])):
        a = note.Note(data_row["Midi numbers"][i], quarterLength = durations[i])
        s.append(a)
    return s
  
def create_notes_without_rhythm_for_render_start_on_c(data_row):
  
    s = stream.Stream()
    for i in range(0, len(data_row["Melodies transposed to start on middle C"])):
        a = note.Note(data_row["Melodies transposed to start on middle C"][i])
        s.append(a)
    return s

def create_notes_with_rhythm_for_renderstart_on_c(data_row):
    durations = data_row["Durations"]
    s = stream.Stream()
    for i in range(0, len(data_row["Melodies transposed to start on middle C"])):
        a = note.Note(data_row["Melodies transposed to start on middle C"][i], quarterLength = durations[i])
        s.append(a)
    return s
  
def create_relative_distances_between_notes(data_row):
    midi_array = data_row["Midi numbers"]
    relative_pitch_array = []
    for i in range(1, len(midi_array)):
        relative_pitch_array.append(midi_array[i] - midi_array[i-1])
    return relative_pitch_array
    
    
def get_number_of_negative_melodic_moves(data_row):
    count = 0
    for i in range(0, len(data_row["Relative distances between notes"])):
        if data_row["Relative distances between notes"][i] < 0:
            count = count + 1
            
    return count
    

def get_number_of_postive_melodic_moves(data_row):
    count = 0
    for i in range(0, len(data_row["Relative distances between notes"])):
        if data_row["Relative distances between notes"][i] > 0:
            count = count + 1
            
    return count

def get_number_of_tones_and_semitones(data_row):
    count = 0
    for i in range(0, len(data_row["Relative distances between notes"])):
        if abs(data_row["Relative distances between notes"][i]) < 3:
            count = count + 1
            
    return count
    

def get_number_moves_min_third_and_above(data_row):
    count = 0
    for i in range(0, len(data_row["Relative distances between notes"])):
        if abs(data_row["Relative distances between notes"][i]) > 2:
            count = count + 1
            
    return count
```

```python
def convert_records_to_musical_phrases(arr):
      
    note_arr = []
    note_array_converted_for_score_arr = []
    temp_note_array = []
    
    full_title_array = []
    temp_title_array = []
    
    measure_array = []
    temp_measure_array = []
    
    location_array = []
    temp_location_array = []
    
    duration_array = []
    temp_duration_array = []
    
    performer_array = []
    temp_performer_array = []
    
    performer_collection_array = []
    temp_performer_collection_array = []
    
    
    for i in range(0, len(arr)):
    
        if arr[i][0] != -1:
            temp_note_array.append(arr[i][0])
            temp_measure_array.append(arr[i][1])
            temp_title_array.append(arr[i][2])
            temp_location_array.append(arr[i][3])
            temp_duration_array.append(arr[i][4])
            temp_performer_array.append(arr[i][5])
            temp_performer_collection_array.append(arr[i][6])
            
        else:
            if len(temp_note_array) > 0:
                note_arr.append(temp_note_array)
                full_title_array.append(temp_title_array)
                measure_array.append(temp_measure_array)
                location_array.append(temp_location_array)
                duration_array.append(temp_duration_array)
                performer_array.append(temp_performer_array)
                performer_collection_array.append(temp_performer_collection_array)
            
            temp_note_array = []
            temp_title_array = []
            temp_measure_array = []
            temp_location_array = []
            temp_duration_array = []
            temp_performer_array = []
            temp_performer_collection_array = []


    return [note_array_converted_for_score_arr, 
            note_arr, 
            full_title_array, 
            measure_array,
            location_array, 
            duration_array,
            performer_array,
            performer_collection_array]
        
```

```python
def create_melody_frame(dataframe,  performer_collection, title, performer="Keith Jarrett"):
    
    data = dataframe[dataframe["Title"] == title]
    data = data[data["Performer name"] == performer]
    data = data[data["Performer collection"] == performer_collection]
    

    
    data = data[["Midi number", 
                 "Current measure", 
                 "Title", 
                 "Location", 
                 "Duration", 
                 "Performer name", 
                 "Performer collection"]].values
    
    melodies = convert_records_to_musical_phrases(data)
    melodies = pd.DataFrame(data=melodies).T
    melodies.columns = ['Render data without rhythm', 
                        'Midi numbers', 
                        'Title', 
                        'Measures', 
                        'Locations', 
                        'Durations', 
                        'Performer name', 
                        'Performer collection']
    
    
    melodies["Title"] = melodies.apply(fix_title, axis=1)
    melodies['Performer name'] = melodies.apply(fix_performer_name, axis=1)
    melodies["Performer collection"] = melodies.apply(fix_performer_collection, axis=1)
    melodies["Phrase length in notes"] = melodies.apply(get_phrase_length, axis=1)
    melodies["Measure that phrase begins"] = melodies.apply(get_opening_measure, axis=1)
    melodies["Measure that phrase ends"] = melodies.apply(get_closing_measure, axis=1)
    melodies["Location in measure that phrase begins"] = melodies.apply(get_starting_position_in_measure, axis=1)
    melodies["Location in measure that phrase ends"] = melodies.apply(get_ending_position_in_measure, axis=1)
    melodies["Music pitch range"] = melodies.apply(get_pitch_range, axis=1)
    melodies["Number of different pitches used"] = melodies.apply(get_number_of_different_pitches_used, axis=1)
    melodies["Highest pitch used"] = melodies.apply(get_highest_pitch, axis=1)
    melodies["Lowest pitch used"] = melodies.apply(get_lowest_pitch, axis =1)

    melodies['Different pitches'] = melodies.apply(get_different_pitches_used, axis=1)
    melodies["Different durations"] = melodies.apply(get_different_durations_used, axis=1)

    melodies["Longest duration"] = melodies.apply(get_longest_duration, axis=1)
    melodies['Shortest duration'] = melodies.apply(get_shortest_duration, axis=1)
    melodies["Number of different durations used in melody"] = melodies.apply(get_number_different_durations_used, axis=1)
    melodies["Melodies transposed to start on middle C"] = melodies.apply(transpose_midi_numbers_to_start_on_c, axis=1)


    melodies["Relative distances between notes"] = melodies.apply(create_relative_distances_between_notes, axis=1)

    melodies["Postive moves in melody"] = melodies.apply(get_number_of_postive_melodic_moves, axis=1)
    melodies["Negative moves in melody"] = melodies.apply(get_number_of_negative_melodic_moves, axis=1)

    melodies["Small melodic moves"] = melodies.apply(get_number_of_tones_and_semitones, axis=1)
    melodies["Big melodic moves"] = melodies.apply(get_number_moves_min_third_and_above, axis=1)

    melodies["Notes without rhythm"] = melodies.apply(create_notes_without_rhythm_for_render, axis=1)
    melodies["Notes with rhythm"] = melodies.apply(create_notes_with_rhythm_for_render, axis=1)

    melodies["Notes without rhythm from C"] = melodies.apply(create_notes_without_rhythm_for_render_start_on_c, axis=1)
    melodies["Notes with rhythm from C"] = melodies.apply(create_notes_with_rhythm_for_renderstart_on_c, axis=1)


#set of pitches
#set of durations

    del melodies["Measures"]
    del melodies["Locations"]
    del melodies["Render data without rhythm"]
    
    
    melodies.columns = [
     'Phrase midi numbers',
     'Title',
     'Phrase durations',
     'Performer name',
     'Performer collection',
     'Number of notes in phrase',
     'Measure in which phrase begins',
     'Measure in which phrase ends',
     'Measure location in which phrase begins',
     'Measure location in which phrase ends',
     'Range of phrase in semitones',
     'Number of pitches used in phrase',
     'Highest pitch used in phrase',
     'Lowest pitch used in phrase',
     'Different pitches used in phrase',
     'Different durations used in phrase',
     'Longest duration used in phrase',
     'Shortest duration used in phrase',
     'Number of different durations used in phrase',
     'Phrase midi numbers transposed to start on middle C',
     'Distances between subsequent phrase notes',
     'Number of positive steps or leaps in phrase',
     'Number of negative steps or leaps in phrase',
     'Number of step movements (by tones or semitones) in phrase',
     'Number of leap movements (by minor thirds or above) in phrase',
     'Phrase without rhythm render data',
     'Phrase with rhythm render data',
     'Phrase without rhythm from C render data',
     'Phrase with rhythm from C render data'
    ]

    return melodies
    
    
kj_dowar_melodies = create_melody_frame(data,
                                        "Keith Jarrett At The Blue Note, The Complete Recordings (Vol. 3)", 
                                        "Days Of Wine And Roses")
kj_sbs_melodies = create_melody_frame(data,
                                     "Standards Live",
                                      "Stella By Starlight")

kj_attya_melodies = create_melody_frame(data,
                                        "Standards, Vol. 1", 
                                        "All The Things You Are")
kj_iliv_melodies = create_melody_frame(data,
                                     "Standards, Vol. 2",
                                      "In Love In Vain")

kj_al1_melodies = create_melody_frame(data,
                                        "Still Live", 
                                        "Autumn Leaves")
kj_mfv_melodies = create_melody_frame(data,
                                     "Still Live",
                                      "My Funny Valentine")

kj_al2_melodies = create_melody_frame(data,
                                        "Tokyo 96", 
                                        "Autumn Leaves")
kj_iiwab_melodies = create_melody_frame(data,
                                     "Up For It",
                                      "If I Were A Bell")

kj_smpwc_melodies = create_melody_frame(data,
                                        "Up For It", 
                                        "Someday My Prince Will Come")
kj_gh_melodies = create_melody_frame(data,
                                     "Whisper Not",
                                      "Groovin High")


```
