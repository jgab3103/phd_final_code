```python
def convert_midi_number_to_string(data_row):
    midi_number = data_row["Midi number"]
    octave = math.floor(midi_number / 12) - 1
    remainder = midi_number % 12
    note_string = ''
    
    if remainder == 0:
        note_string = 'C'
    elif remainder == 1:
        note_string = 'C#/Db'
    elif remainder == 2:
        note_string = 'D'
    elif remainder == 3:
        note_string = 'D#/Eb'
    elif remainder == 4:
        note_string = 'E'
    elif remainder == 5:
        note_string = 'F'
    elif remainder == 6:
        note_string = 'F#/Gb'
    elif remainder == 7:
        note_string = 'G'
    elif remainder == 8:
        note_string = 'G#/Ab'
    elif remainder == 9:
        note_string = 'A'
    elif remainder == 10:
        note_string = 'A#/Bb'
    elif remainder == 11:
        note_string = 'B'
    
    
    
    if data_row["Midi number"] < 0:
        return 'rest'
    else:
        return note_string + str(octave)
    
    
def convert_midi_number_to_string_without_octave(data_row):
    midi_number = data_row["Midi number"]
    octave = math.floor(midi_number / 12) - 1
    remainder = midi_number % 12
    note_string = ''
    
    if remainder == 0:
        note_string = 'C'
    elif remainder == 1:
        note_string = 'C#/Db'
    elif remainder == 2:
        note_string = 'D'
    elif remainder == 3:
        note_string = 'D#/Eb'
    elif remainder == 4:
        note_string = 'E'
    elif remainder == 5:
        note_string = 'F'
    elif remainder == 6:
        note_string = 'F#/Gb'
    elif remainder == 7:
        note_string = 'G'
    elif remainder == 8:
        note_string = 'G#/Ab'
    elif remainder == 9:
        note_string = 'A'
    elif remainder == 10:
        note_string = 'A#/Bb'
    elif remainder == 11:
        note_string = 'B'
    
    
    
    if data_row["Midi number"] < 0:
        return 'rest'
    else:
        return note_string
    

    
def convert_float_duration_to_string(data_row):
    if data_row['Duration due to tied notes'] == 240:
        return 'Eighth note'
    elif data_row['Duration due to tied notes'] == 120:
        return "Sixteenth note"
    elif data_row['Duration due to tied notes'] == 160:
        return "Twelth note"
    elif data_row['Duration due to tied notes'] == 480:
        return "Quarter note"
    elif data_row['Duration due to tied notes'] == 0:
        return "Rest"
    elif data_row['Duration due to tied notes'] == 960:
        return "Half note"
    elif data_row['Duration due to tied notes'] == 720:
        return "Dotted quarter note"
    elif data_row['Duration due to tied notes'] == 80:
        return "Thirty-sixth note"
    elif data_row['Duration due to tied notes'] == 96:
        return "Twentieth note"
    elif data_row['Duration due to tied notes'] == 320:
        return "Sixth note"
    elif data_row['Duration due to tied notes'] == 360:
        return "Dotted eighth note"
    elif data_row['Duration due to tied notes'] == 1920:
        return "Whole note"
    elif data_row['Duration due to tied notes'] == 60:
        return "Thirty second note"
    elif data_row['Duration due to tied notes'] == 1200:
        return "Half note tied to an eighth note"
    elif data_row['Duration due to tied notes'] == 137:
        return "Fourteenth note"
    elif data_row['Duration due to tied notes'] == 1440:
        return "Dotted half note"
    elif data_row['Duration due to tied notes'] == 192:
        return "Tenth note"
    elif data_row['Duration due to tied notes'] == 107:
        return "Eighteenth note"
    elif data_row['Duration due to tied notes'] ==400:
        return "Twelth note tied to eighth note"
    
    elif data_row['Duration due to tied notes'] ==600:
        return "Quarter note tied to sixteenth note"
    elif data_row['Duration due to tied notes'] ==640:
        return "Quarter note tied to twelth note"
    elif data_row['Duration due to tied notes'] ==180:
        return "Dotted sixteenth note"
    elif data_row['Duration due to tied notes'] ==2160:
        return "Eighth note tied to a whole note"
    elif data_row['Duration due to tied notes'] ==40:
        return "Forty-eighth note"
    elif data_row['Duration due to tied notes'] ==560:
        return "Quarter note tied to a thirty-sixth note"
    elif data_row['Duration due to tied notes'] ==280:
        return "Sixteenth tied to a twelth note"
    
    elif data_row['Duration due to tied notes'] ==200:
        return "Sixteenth note tied to a twenty fourth note"
    elif data_row['Duration due to tied notes'] ==800:
        return "Quarter note tied to a sixth note"
    elif data_row['Duration due to tied notes'] ==1080:
        return "Half note tied to a sixteenth note"
    elif data_row['Duration due to tied notes'] ==1800:
        return "Half note tied to a dotted eighth note"
    elif data_row['Duration due to tied notes'] ==300:
        return "Eighth note tied to a thirty-second note"
    elif data_row['Duration due to tied notes'] ==2400:
        return "Whole note tied to a quarter note"
    elif data_row['Duration due to tied notes'] ==2880:
        return "Whole note tied to a half note"
    elif data_row['Duration due to tied notes'] ==840:
        return "Quarter note tied to a dotted eighth note"
    elif data_row['Duration due to tied notes'] ==880:
        return "Third note tied to a sixth note"
    elif data_row['Duration due to tied notes'] ==1040:
        return "Sixth note tied to a dotted quarter note"
    else:
        return 'fix'
```
