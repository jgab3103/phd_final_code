```python
all_melodies = pd.read_pickle('./data_frames/all_melodies_frame')
```

```python
def create_stream(data_row):
    data = eval(data_row["Phrase details in midi numbers"])
    s = stream.Stream()
    for i in range(0, len(data)):
        a = note.Note(data[i])
        s.append(a)
        
    return s
    
```

```python
# LOOK AT UNIQUE MELODIES
# Confounding, there is not much repetition, which makes jarrett so hard to deal with

unique_melodies = all_melodies
unique_melodies['_str_ Melody midi numbers'] = unique_melodies["Phrase midi numbers"].astype(str)
unique_melodies = unique_melodies[unique_melodies["Performer name"] == "Keith Jarrett"]
unique_kj_melodies = unique_melodies["_str_ Melody midi numbers"].value_counts().to_frame().reset_index()
unique_kj_melodies.columns = ['Phrase details in midi numbers', 'Number of times phrase occurs']
unique_kj_melodies = unique_kj_melodies[unique_kj_melodies['Number of times phrase occurs'] > 1]
unique_kj_melodies["Render data"] = unique_kj_melodies.apply(create_stream, axis=1)
```
