```python
styles = {
    'graph': {
        'label': 'Probability of the choosing the a given pitch class following C',
        'fontsize': '10',
        'fontname': 'arial',
        'fontcolor': 'dark gray',
        'bgcolor': '#E5E5E5',
        #'rankdir': 'BT',
    },
    'nodes': {
        'fontname': 'arial',
       # 'shape': 'box',
        'fontcolor': 'white',
        'fontsize':'8',
        'color': 'white',
        'style': 'filled',
        'fillcolor': '#E24A33',
    },
    'edges': {
        #'style': 'dashed',
        'color': 'dark gray',
        'arrowhead': 'open',
        'fontname': 'Courier',
        'fontsize': '8',
        'fontcolor': 'white',
    }
}

def apply_styles(graph, styles):
    graph.graph_attr.update(
        ('graph' in styles and styles['graph']) or {}
    )
    graph.node_attr.update(
        ('nodes' in styles and styles['nodes']) or {}
    )
    graph.edge_attr.update(
        ('edges' in styles and styles['edges']) or {}
    )
    return graph
```

```python


dot = Digraph()


dot.node('A', 'C')
dot.node('B', 'C#/Db\n18%')
dot.node('C', 'D\n22%')
dot.node('D', 'D#/Eb\n12%')
dot.node('E', 'E\n8%')
dot.node('F', 'F\n2%')
dot.node('G', 'F#/Gb\n16%')
dot.node('H', 'G\n28%')
dot.node('I', 'G#/Ab\n23%')
dot.node('J', 'A\n16%')
dot.node('K', 'A#/Bb\n28%')
dot.node('L', 'B\n52%')


dot.edges(['AB', 'AC', 'AD', 'AE', 'AF', 'AG', 'AH', 'AI', 'AJ', 'AK', 'AL'])


dot = apply_styles(dot, styles)
dot
```

```python
two_note_microphrases = pd.read_pickle('../Datasets/DataFrames/two_note_microphrases')
```

```python
two_note_microphrases["Microphase midi numbers STR"] = two_note_microphrases["Microphase midi numbers"].astype(str)
two_note_microphrases["Microphrase durations STR"] = two_note_microphrases["Microphrase durations"].astype(str)
two_note_microphrases['Microphrase midi numbers transposed to start on middle C STR'] = two_note_microphrases["Microphrase midi numbers transposed to start on middle C"].astype(str)
two_note_microphrases['Relative distances between microphrase notes STR'] = two_note_microphrases['Relative distances between microphrase notes'].astype(str)
two_note_microphrases["Microphrase midi numbers as note names STR"] = two_note_microphrases["Microphrase midi numbers as note names"].astype(str)
```

```python
two_note_kj = two_note_microphrases[two_note_microphrases["Performer name"] == "Keith Jarrett"]
two_note_r = two_note_microphrases[two_note_microphrases["Performer name"] != "Keith Jarrett"]
```

```python
#strategy
#number of times microphrase appears in KJ and then ref solo
# examples from both ends and middle, 45 vs 20 vs 3
# start with part of string - percentage chance
# number of times relative appears in KJ and then ref solo
# look at how many start with part of relative string
# order of what is likely to happen after first number
# very unlikely scenarios
# similiar based on tonality
# at each note, probablility, of what will happen next
```

```python
x = two_note_kj[two_note_kj["Microphrase midi numbers as note names STR"].str.startswith("['A#/Bb4")]
x = x.groupby('Microphrase midi numbers as note names STR').count()["Microphrase midi numbers as note names"].to_frame()
x = x.sort_values('Microphrase midi numbers as note names', ascending=False)
x["Percentage"] = (x["Microphrase midi numbers as note names"] / x["Microphrase midi numbers as note names"].sum()) * 100
x['Percentage'].to_frame().T.plot(kind="barh", stacked=True, figsize=(15,2), title="Distribution of notes")
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```

```python
two_note_kj["Relative distances between microphrase notes STR"].value_counts().to_frame().reset_index(drop=True).plot()
```

```python
x = two_note_kj[two_note_kj["Microphase midi numbers STR"].str.startswith('[74')]
x = x.groupby('Microphase midi numbers STR').count()["Microphase midi numbers"].to_frame()
x = x.sort_values('Microphase midi numbers', ascending=False)
x["Perc"] = (x["Microphase midi numbers"] / x["Microphase midi numbers"].sum()) * 100
x['Perc'].to_frame().T.plot(kind="barh", stacked=True, figsize=(15,2), title="Distribution of notes")
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```
