```python
# NEW FUNCTION TO BREAK PHRASES INTO MICROPHRASES



def convert_phrase_to_microphrases(data):
    microphrase_array = []
    #print(data)

    for i in range(0, len(data)):
        phrase_array = data[i][0]
        duration_array = data[i][2]
        
        
        
        for j in range(0, len(phrase_array)):
            temp_microphrase_array = []
            if len(phrase_array[j:j+MICROPHRASE_SIZE])== MICROPHRASE_SIZE:
                temp_microphrase_array.append(phrase_array[j:j+MICROPHRASE_SIZE])
                temp_microphrase_array.append(data[i][1])
                temp_microphrase_array.append(duration_array[j:j+MICROPHRASE_SIZE])
                temp_microphrase_array.append(data[i][3])
                temp_microphrase_array.append(data[i][4])
                temp_microphrase_array.append(data[i][5])
                temp_microphrase_array.append(data[i][6])
                temp_microphrase_array.append(data[i][7])
                temp_microphrase_array.append(data[i][8])

            if len(temp_microphrase_array) > 0:
                microphrase_array.append(temp_microphrase_array)
                
                
        

    
    return microphrase_array
         
        
```

```python
#FUNCTIONS TO BUILD PHRASE DATA
def get_phrase_length(data_row):
    return len(data_row["Midi numbers"])

def get_pitch_range(data_row):
    return max(data_row["Midi numbers"]) - min(data_row["Midi numbers"])


def get_number_of_different_pitches_used(data_row):
    return len(set(data_row["Midi numbers"]))

def get_highest_pitch(data_row):
    return max(data_row["Midi numbers"])

def get_lowest_pitch(data_row):
    return  min(data_row["Midi numbers"])

def get_number_of_different_pitches_used(data_row):
    return len(set(data_row["Midi numbers"]))

def get_different_pitches_used(data_row):
    return list(set(data_row["Midi numbers"]))

def get_longest_duration(data_row):
    return max(data_row["Durations"])

def get_shortest_duration(data_row):
    return min(data_row["Durations"])

def get_number_different_durations_used(data_row):
    return len(set(data_row["Durations"]))

def get_different_durations_used(data_row):
    return list(set(data_row["Durations"]))


def transpose_midi_numbers_to_start_on_c(data_row):
    transposed_array = []
    first_note_in_melody = data_row["Midi numbers"][0]
    for i in range(0, len(data_row["Midi numbers"])):
        transposed_array.append(data_row["Midi numbers"][i] - first_note_in_melody + 60)
        
    return transposed_array


def create_notes_without_rhythm_for_render(data_row):
  
    s = stream.Stream()
    for i in range(0, len(data_row["Midi numbers"])):
        a = note.Note(data_row["Midi numbers"][i])
        s.append(a)
    return s

def create_notes_with_rhythm_for_render(data_row):
    durations = data_row["Durations"]
    s = stream.Stream()
    for i in range(0, len(data_row["Midi numbers"])):
        a = note.Note(data_row["Midi numbers"][i], quarterLength = durations[i])
        s.append(a)
    return s
  
def create_notes_without_rhythm_for_render_start_on_c(data_row):
  
    s = stream.Stream()
    for i in range(0, len(data_row["Melodies transposed to start on middle C"])):
        a = note.Note(data_row["Melodies transposed to start on middle C"][i])
        s.append(a)
    return s

def create_notes_with_rhythm_for_renderstart_on_c(data_row):
    durations = data_row["Durations"]
    s = stream.Stream()
    for i in range(0, len(data_row["Melodies transposed to start on middle C"])):
        a = note.Note(data_row["Melodies transposed to start on middle C"][i], quarterLength = durations[i])
        s.append(a)
    return s
  
def create_relative_distances_between_notes(data_row):
    midi_array = data_row["Midi numbers"]
    relative_pitch_array = []
    for i in range(1, len(midi_array)):
        relative_pitch_array.append(midi_array[i] - midi_array[i-1])
    return relative_pitch_array
    
    
def get_number_of_negative_melodic_moves(data_row):
    count = 0
    for i in range(0, len(data_row["Relative distances between notes"])):
        if data_row["Relative distances between notes"][i] < 0:
            count = count + 1
            
    return count
    

def get_number_of_postive_melodic_moves(data_row):
    count = 0
    for i in range(0, len(data_row["Relative distances between notes"])):
        if data_row["Relative distances between notes"][i] > 0:
            count = count + 1
            
    return count

def get_number_of_tones_and_semitones(data_row):
    count = 0
    for i in range(0, len(data_row["Relative distances between notes"])):
        if abs(data_row["Relative distances between notes"][i]) < 3:
            count = count + 1
            
    return count
    

def get_number_moves_min_third_and_above(data_row):
    count = 0
    for i in range(0, len(data_row["Relative distances between notes"])):
        if abs(data_row["Relative distances between notes"][i]) > 2:
            count = count + 1
            
    return count

def convert_midi_number_to_string(midi_number):
    
    midi_number = midi_number
    octave = math.floor(midi_number / 12) - 1
    remainder = midi_number % 12
    note_string = ''
    
    if remainder == 0:
        note_string = 'C'
    elif remainder == 1:
        note_string = 'C#/Db'
    elif remainder == 2:
        note_string = 'D'
    elif remainder == 3:
        note_string = 'D#/Eb'
    elif remainder == 4:
        note_string = 'E'
    elif remainder == 5:
        note_string = 'F'
    elif remainder == 6:
        note_string = 'F#/Gb'
    elif remainder == 7:
        note_string = 'G'
    elif remainder == 8:
        note_string = 'G#/Ab'
    elif remainder == 9:
        note_string = 'A'
    elif remainder == 10:
        note_string = 'A#/Bb'
    elif remainder == 11:
        note_string = 'B'
    
    
    
    if midi_number < 0:
        return 'rest'
    else:
        return note_string + str(octave)

    
    
def create_note_string(data_row):
    s = []
    for i in range(0, len(data_row["Midi numbers"])):
        a = convert_midi_number_to_string(data_row["Midi numbers"][i])
        s.append(a)
    return s
```

```python
#CREATE ALL OTHER AGGREGATIONS
all_microphrases["Microphrase midi numbers as letters"] = all_microphrases.apply(create_note_string, axis=1)
all_microphrases["Phrase length in notes"] = all_microphrases.apply(get_phrase_length, axis=1)
all_microphrases["Music pitch range"] = all_microphrases.apply(get_pitch_range, axis=1)
all_microphrases["Number of different pitches used"] = all_microphrases.apply(get_number_of_different_pitches_used, axis=1)
all_microphrases["Highest pitch used"] = all_microphrases.apply(get_highest_pitch, axis=1)
all_microphrases["Lowest pitch used"] = all_microphrases.apply(get_lowest_pitch, axis =1)
all_microphrases['Different pitches'] = all_microphrases.apply(get_different_pitches_used, axis=1)
all_microphrases["Different durations"] = all_microphrases.apply(get_different_durations_used, axis=1)
all_microphrases["Longest duration"] = all_microphrases.apply(get_longest_duration, axis=1)
all_microphrases['Shortest duration'] = all_microphrases.apply(get_shortest_duration, axis=1)
all_microphrases["Number of different durations used in melody"] = all_microphrases.apply(get_number_different_durations_used, axis=1)
all_microphrases["Melodies transposed to start on middle C"] = all_microphrases.apply(transpose_midi_numbers_to_start_on_c, axis=1)
all_microphrases["Relative distances between notes"] = all_microphrases.apply(create_relative_distances_between_notes, axis=1)
all_microphrases["Postive moves in melody"] = all_microphrases.apply(get_number_of_postive_melodic_moves, axis=1)
all_microphrases["Negative moves in melody"] = all_microphrases.apply(get_number_of_negative_melodic_moves, axis=1)
all_microphrases["Small melodic moves"] = all_microphrases.apply(get_number_of_tones_and_semitones, axis=1)
all_microphrases["Big melodic moves"] = all_microphrases.apply(get_number_moves_min_third_and_above, axis=1)
all_microphrases["Notes without rhythm"] = all_microphrases.apply(create_notes_without_rhythm_for_render, axis=1)
all_microphrases["Notes with rhythm"] = all_microphrases.apply(create_notes_with_rhythm_for_render, axis=1)
all_microphrases["Notes without rhythm from C"] = all_microphrases.apply(create_notes_without_rhythm_for_render_start_on_c, axis=1)
all_microphrases["Notes with rhythm from C"] = all_microphrases.apply(create_notes_with_rhythm_for_renderstart_on_c, axis=1)
```
