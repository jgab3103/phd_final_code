```python
# the theme of variation still in play
# It is also possible examine locations, starting locations
# shows phrase vehicle of variety - enforces previous analysis
# EXAMINE PHRASE STARTING LOCATIONS IN MEASURE ACROSS KJ SOLOS
# This challenges what we know - don't play on the on the one, jarrett doesn't do this, 
# allthough the two is is 

ax = all_melodies["Measure location in which phrase begins"].hist(bins=50)
ax.set_title('Phrase starting locations within measures across all solos', fontsize=12)
ax.set_xlabel("Measure location in which phrase begins")
ax.set_ylabel(labels[0])
```
